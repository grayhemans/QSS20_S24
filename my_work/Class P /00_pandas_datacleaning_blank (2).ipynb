```python
df1 = df.groupby("WARD")["OFFENSE"].unique()
def count_fx(x):
    return len(x)

def join_text(x):
    return ";".join(x)

new_list =[]
for i, l in list (df1.items()):
    new_list.append(join_text(l))

df1.apply(count_fx)
df2 = df1.reset_index()
df2["unqiue_off_by_ward"] = new_list 
df2
                   
```

### map.recode

**Task**: recode shifts that are MIDNIGHT or EVENING as "nighttime"; code other shift to daytime

```python
SHIFTS = ["MIDNIGHT", "EVENING", "DAY"]

new_period = ["nighttimes", "nighttimes", "daytime"]
convert_dict = dict(zip(SHIFTS, new_period))

df["SHIFT_NEW"] = df.SHIFT.map(convert_dict) 
df.SHIFT_NEW.unique()
```

```python
cat = ["nighttime"]
condition = [df.SHIFT.isin(["MIDNIGHT", "EVENING"])]
df["SHIFT_RECODE"] = np.select(condition, cat, default = "daytime")
df.SHIFT_RECODE.values_counts()
```
