**One major caveat**: when investigating whether two similar defendants received different sentences, we're missing one important attribute that influences sentencing: the defendant's criminal history. This influences sentencing both through sentencing guidelines, which can prescribe longer sentences for those who have certain types of prior convictions, and through judicial discretion if judges are more lenient with first-time defendants. The above sources discuss how much we want to "control away" for this prior history, since if we think there are racial biases in which defendants, conditional on *committing* a crime, are arrested and charged, we may not want to adjust for that factor. More discussion [in this article](https://www.themarshallproject.org/2019/12/03/the-growing-racial-disparity-in-prison-time)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def calculate_disparity(dataframe):
    """Calculate the disparity in incarceration rates between white and black defendants."""
    total_white = len(dataframe[dataframe['is_white_derived']])
    total_black = len(dataframe[dataframe['is_black_derived']])
    incarcerated_white = len(dataframe[(dataframe['is_white_derived']) & (dataframe['COMMITMENT_TYPE'] == "Illinois Department of Corrections")])
    incarcerated_black = len(dataframe[(dataframe['is_black_derived']) & (dataframe['COMMITMENT_TYPE'] == "Illinois Department of Corrections")])
    rate_white = incarcerated_white / max(total_white, 1)
    rate_black = incarcerated_black / max(total_black, 1)
    return rate_white - rate_black

def calculate_standard_error(rate_white, rate_black, total_white, total_black):
    """Calculate the standard error for the disparity."""
    p = abs(rate_white - rate_black)
    n = total_white + total_black
    return np.sqrt(p * (1 - p) / max(n, 1))
```

```python
# Filtering data for retail theft charges
df_retail_theft = df_clean[df_clean['simplified_offense_derived'] == "Retail Theft"]

# Time frames for analysis
time_frames = [
    ("2016-10-01", "2016-11-30", "2017-01-01", "2017-02-28"),
    ("2016-08-01", "2016-11-30", "2017-01-01", "2017-04-30"),
    ("2016-04-01", "2016-11-30", "2017-01-01", "2017-08-31"),
    ("2015-11-01", "2016-11-30", "2017-01-01", "2017-12-31")
]

# Preparing the results DataFrame
results = pd.DataFrame(columns=["Bandwidth", "Before", "After", "Standard Error Before", "Standard Error After"])

for index, (start_pre, end_pre, start_post, end_post) in enumerate(time_frames):
    pre_period = df_retail_theft[(df_retail_theft['sentenceym_derived'] >= start_pre) & (df_retail_theft['sentenceym_derived'] <= end_pre)]
    post_period = df_retail_theft[(df_retail_theft['sentenceym_derived'] >= start_post) & (df_retail_theft['sentenceym_derived'] <= end_post)]
    
    disparity_pre = calculate_disparity(pre_period)
    disparity_post = calculate_disparity(post_period)
    
    error_pre = calculate_standard_error(disparity_pre, disparity_pre, len(pre_period[pre_period['is_white_derived']]), len(pre_period[pre_period['is_black_derived']]))
    error_post = calculate_standard_error(disparity_post, disparity_post, len(post_period[post_period['is_white_derived']]), len(post_period[post_period['is_black_derived']]))
    
    results.loc[index] = [index+1, disparity_pre, disparity_post, error_pre, error_post]

print(results)
```
